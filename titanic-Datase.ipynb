```python
import matplotlib.pyplot as plt
import seaborn as sns


num_cols = ['Age', 'Fare', 'SibSp', 'Parch']

plt.figure(figsize=(12,8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)   # 2 rows, 2 cols
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]

for col in num_cols:
    df = remove_outliers_iqr(df, col)

print("After removing outliers, shape:", df.shape)
```

```python
plt.figure(figsize=(12,8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col} after outlier removal')

plt.tight_layout()
plt.show()
```

```python
def remove_outliers_iqr(data, cols):
    for col in cols:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        data = data[(data[col] >= lower) & (data[col] <= upper)]
    return data

num_cols = ['Age', 'Fare', 'SibSp', 'Parch']
df_clean = remove_outliers_iqr(df, num_cols)

print("Before:", df.shape)
print("After :", df_clean.shape)
```

```python
from scipy.stats.mstats import winsorize

# Example: Winsorize 'Fare' at 5% low, 95% high
df['Fare'] = winsorize(df['Fare'], limits=[0.05, 0.05])

# Same for Age
df['Age'] = winsorize(df['Age'], limits=[0.05, 0.05])
```
